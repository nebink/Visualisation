```python
FuelType = {"Petrol": 1, "Diesel": 2}
Data['FuelType'] = Data['FuelType'].map(FuelType)
```

```python
numerical_vars = Data.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_vars = Data.select_dtypes(include=['object']).columns.tolist()                           
print('Numerical variables:', numerical_vars)
print('Categorical variables:', categorical_vars)
```

```python
Transmission = {"Manual": 1, "Automatic": 2}
Data['Transmission'] = Data['Transmission'].map(Transmission)
```

```python
numerical_vars = Data.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_vars = Data.select_dtypes(include=['object']).columns.tolist()                           
print('Numerical variables:', numerical_vars)
print('Categorical variables:', categorical_vars)
```

```python
# Count occurrences of each FuelType
fuel_counts = df['FuelType'].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(fuel_counts, labels=("Petrol","Diesel"), autopct='%1.1f%%', startangle=140, colors=['gold', 'lightcoral', 'lightskyblue'])
plt.title('Distribution of Cars by FuelType')
plt.axis('equal')
plt.show()
```

```python
# Count occurrences of each Transmission type
transmission_counts = df['Transmission'].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(transmission_counts, labels=("Manual","Automatic"), autopct='%1.1f%%', startangle=140, colors=['gold', 'lightcoral'])
plt.title('Distribution of Cars by Transmission Type')
plt.axis('equal')
plt.show()
```

```python
# Histogram of Prices
plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], bins=10, kde=True, color='blue')
plt.title('Distribution of Car Prices')
plt.xlabel('Price (in Lakh)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Histogram of Kilometers Driven
plt.figure(figsize=(10, 6))
sns.histplot(df['KilometersDriven'], bins=10, kde=True, color='green')
plt.title('Distribution of Kilometers Driven')
plt.xlabel('Kilometers Driven')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot of Prices by FuelType
plt.figure(figsize=(10, 6))
sns.boxplot(x='FuelType', y='Price', data=df)
plt.title('Box Plot of Prices by FuelType')
plt.xlabel('FuelType')
plt.ylabel('Price (in Lakh)')
plt.show()
```

```python
# Scatter plot of Kilometers Driven vs Price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='KilometersDriven', y='Price', hue='FuelType', style='Transmission', data=df)
plt.title('Scatter Plot of Kilometers Driven vs Price')
plt.xlabel('Kilometers Driven')
plt.ylabel('Price (in Lakh)')
plt.legend()
plt.show()
```

```python
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
```

```python
avg_price_by_year = df.groupby('YearOfManufacture')['Price'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='YearOfManufacture', y='Price', data=avg_price_by_year, palette='viridis')
plt.title('Average Price by Year of Manufacture')
plt.xlabel('Year of Manufacture')
plt.ylabel('Average Price (in Lakh)')
plt.show()
```

```python
# Pair plot to see relationships between multiple pairs of columns
sns.pairplot(df, hue='FuelType', markers=["o", "s", "D", "^"])
plt.suptitle('Pair Plot for Various Columns', y=1.02)
plt.show()
```
